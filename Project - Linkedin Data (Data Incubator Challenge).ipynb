```python
# Plotting a Histogram
Linkedindata.company_name.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('company_name')
plt.ylabel('followers_count')
plt.xlabel('company_name');
```

```python
# Finding the relations between the variables.
plt.figure(figsize=(20,10))
cor= Linkedindata.corr()
cor
```
